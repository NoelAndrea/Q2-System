```python
from datasets import load_dataset
from pathlib import Path
from PIL import Image
import os

# Create base directory
base_dir = Path("data")
base_dir.mkdir(exist_ok=True)

# Class mapping (0,1,2 → folder names)
label_to_name = {
    0: "angular_leaf_spot",
    1: "bean_rust",
    2: "healthy"
}

# Load each split separately to preserve the original train/val/test division
splits = {
    "train": load_dataset("beans", split="train"),
    "validation": load_dataset("beans", split="validation"),
    "test": load_dataset("beans", split="test"),
}

# Save images to folder structure
for split_name, split_dataset in splits.items():
    print(f"Saving {split_name} split... ({len(split_dataset)} images)")

    for i, example in enumerate(split_dataset):
        img: Image.Image = example["image"]           # This is a PIL Image
        label_idx = example["labels"]                 # integer 0, 1 or 2

        class_name = label_to_name[label_idx]
        class_folder = base_dir / split_name / class_name

        # Create class folder if it doesn't exist
        class_folder.mkdir(parents=True, exist_ok=True)

        # Save image with numbered name
        save_path = class_folder / f"img_{i:04d}.jpg"
        img.save(save_path)

print("\nDataset successfully saved!")
print("Folder structure should now look like:")
print("data/")
print("├── train/")
print("│   ├── angular_leaf_spot/")
print("│   ├── bean_rust/")
print("│   └── healthy/")
print("├── validation/  (same subfolders)")
print("└── test/        (same subfolders)")

# Quick check
print("\nTrain classes:")
!ls -l data/train

print("\nExample counts:")
!find data/train -type f | wc -l
!find data/validation -type f | wc -l
!find data/test -type f | wc -l
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import os
import copy

# === CONFIG (easy to change) ===
data_dir = 'data'
batch_size = 32
num_epochs = 15
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === DATA TRANSFORMS (preparation) ===
data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),      # random zoom
        transforms.RandomHorizontalFlip(),      # mirror image
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
}

# Load images automatically from folders
image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x), data_transforms[x]) for x in ['train', 'val']}
dataloaders = {x: DataLoader(image_datasets[x], batch_size=batch_size, shuffle=True, num_workers=2) for x in ['train', 'val']}
class_names = image_datasets['train'].classes
print("Classes:", class_names)

# === TRANSFER LEARNING MODEL ===
model = models.resnet50(pretrained=True)

# Freeze early layers (they already know shapes)
for param in model.parameters():
    param.requires_grad = False

# Change only the last layer for our 3 classes
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(class_names))
model = model.to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# Unfreeze last block so it can learn bean-specific features
for param in model.layer4.parameters():
    param.requires_grad = True

# === TRAINING LOOP ===
def train_model(model, criterion, optimizer, num_epochs=15):
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)
                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / len(image_datasets[phase])
            epoch_acc = running_corrects.double() / len(image_datasets[phase])
            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())

    model.load_state_dict(best_model_wts)
    return model

# START TRAINING
model = train_model(model, criterion, optimizer, num_epochs)
torch.save(model.state_dict(), 'bean_disease_model.pth')
print("✅ Training finished! Model saved as bean_disease_model.pth")
```

```python
test_transform = data_transforms['val']
test_dataset = datasets.ImageFolder('data/test', test_transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

model.eval()
correct = total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Test Accuracy: {100 * correct / total:.2f}%')
```

```python
from PIL import Image
import torch.nn.functional as F

def predict(image_path):
    transform = data_transforms['val']
    img = Image.open(image_path).convert('RGB')
    img_tensor = transform(img).unsqueeze(0).to(device)

    model.eval()
    with torch.no_grad():
        outputs = model(img_tensor)
        probs = F.softmax(outputs, dim=1)
        confidence, pred = torch.max(probs, 1)

    disease = class_names[pred.item()]
    print(f"🌱 Diagnosis: **{disease}** (Confidence: {confidence.item():.1%})")
    return disease, confidence.item()

# Test it (first upload your photo to Colab left sidebar)
predict('/content/bean_leaf.jpg')   # change to your file name
```

```python
import gradio as gr

def classify(img):
    img.save("temp.jpg")
    pred, conf = predict("temp.jpg")
    return f"**{pred}** (Confidence: {conf:.1%})"

iface = gr.Interface(
    fn=classify,
    inputs=gr.Image(type="pil", label="📸 Upload Bean Leaf Photo"),
    outputs=gr.Label(label="Diagnosis"),
    title=" KWATHU FARMS AI – Bean Disease Detector 🌾",
    description="Upload a leaf photo → Get instant diagnosis: Helping Farmers Protect Their Crops",
    examples=["data/test/healthy/any.jpg", "data/test/angular_leaf_spot/any.jpg", "data/test/bean_rust/any.jpg"]
)
iface.launch(share=True)
```
